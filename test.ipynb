```python
from hkrequests import HKRequests
from pdf_analysis import PdfAnalysis

from datetime import datetime, timedelta
from pprint import pprint
from typing import *

from transformers import BartForConditionalGeneration
from transformers import PreTrainedTokenizerFast
```

```python
start_date = str((datetime.now() - timedelta(days=20)).date())
end_date = str(datetime.now().date())

model = BartForConditionalGeneration.from_pretrained('gogamza/kobart-summarization').eval()
tokenizer = PreTrainedTokenizerFast.from_pretrained('gogamza/kobart-summarization')
```

```python
try:
    analyzer = PdfAnalysis(_model=model, _tokenizer=tokenizer)
    crawler = HKRequests(_sdate=start_date, _edate=end_date, _analyzer=analyzer)
    reports: List[Dict] = crawler.request()

except AssertionError as e:
    print(e)

print(len(reports))
```

```python
# result
for num, report in enumerate(reports):
    pprint(num + 1)
    pprint(report)
```
