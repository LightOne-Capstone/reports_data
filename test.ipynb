```python
from hkrequests import HKRequests
```

```python
from datetime import datetime, timedelta
from pprint import pprint
from typing import *
```

```python
start_date = str((datetime.now() - timedelta(days=10)).date())
end_date = str(datetime.now().date())

try:
    crawler = HKRequests(_sdate=start_date, _edate=end_date)
    reports: List[Dict] = crawler.request()

except AssertionError as e:
    print(e)

print(len(reports))
```

```python
# result
for report in reports:
    pprint(report)
```
